# Pergunta 1.   
   ## Qual a quantidade de escolas que não possuem acesso a internet para utilização de ensino e aprendizagem? (Escolas em geral)

```python
spark.sql('''
        SELECT
            nu_ano_censo,
            COUNT(1) as total_escolas_publicas_sem_internet_aprendizagem
        FROM microdadados_tecnologia_2020_2022
        WHERE in_internet_aprendizagem == 0
        GROUP BY nu_ano_censo

''')
```

# Pergunta 2.

```python
spark.sql('''
        SELECT
            nu_ano_censo,
            COUNT(1) as total_escolas_publicas_sem_internet_aprendizagem
        FROM microdadados_tecnologia_2020_2022
        WHERE tp_dependencia != 4
        AND in_internet_aprendizagem == 0
        GROUP BY nu_ano_censo

''')
```

```python
spark.sql('''
        SELECT
            nu_ano_censo,
            COUNT(1) as total_escolas_privadas_sem_internet_aprendizagem
        FROM microdadados_tecnologia_2020_2022
        WHERE tp_dependencia == 4
        AND in_internet_aprendizagem == 0
        GROUP BY nu_ano_censo

''')
```

# Pergunta 4.   
   ## Quantas escolas não possuem computadores (Desktop, portátil e tablet.)? (dividir escolas publicas e privadas)

```python
spark.sql('''
        SELECT
            nu_ano_censo,
            COUNT(1) as total_escolas_publicas_sem_computadores
        FROM microdadados_tecnologia_2020_2022
        WHERE tp_dependencia != 4
        AND in_desktop_aluno == 0
        AND in_comp_portatil_aluno == 0
        AND in_tablet_aluno == 0
        GROUP BY nu_ano_censo

''')
```

```python
spark.sql('''
        SELECT
            nu_ano_censo,
            COUNT(1) as total_escolas_privadas_sem_computadores
        FROM microdadados_tecnologia_2020_2022
        WHERE tp_dependencia == 4
        AND in_desktop_aluno == 0
        AND in_comp_portatil_aluno == 0
        AND in_tablet_aluno == 0
        GROUP BY nu_ano_censo

''')
```

# Pergunta 5.   
   ## Qual a média de computadores (Desktop, portátil e tablet.) disponiveis para os alunos nas escolas em geral?  

```python
spark.sql('''
        SELECT
            nu_ano_censo,
            ROUND(AVG(qt_desktop_aluno), 0) AS media_desktop_alunos,
            ROUND(AVG(qt_comp_portatil_aluno), 0) AS media_portatil_alunos,
            ROUND(AVG(qt_tablet_aluno), 0) AS media_tablet_alunos
        FROM microdadados_tecnologia_2020_2022
        WHERE qt_desktop_aluno IS NOT null
        AND qt_desktop_aluno < 88888 
        AND qt_comp_portatil_aluno IS NOT null
        AND qt_comp_portatil_aluno < 88888
        AND qt_tablet_aluno IS NOT null
        AND qt_tablet_aluno < 88888
        GROUP BY nu_ano_censo

''')
```
